```python
# Tedarikçilere göre Sahte Parça (Fiziksel) kontrol sonucu dağılımı – Pivot Table + Heatmap

# 1. Pivot tablo oluştur: Tedarikçi & Sahte Parça (Fiziksel) sonuçları
pivot_sp = pd.pivot_table(df,
                          index="Tedarikçi",
                          columns="Sahte Parça (Fiziksel)",
                          aggfunc="size",
                          fill_value=0)

# 2. En çok veri gönderen 10 tedarikçiyi seç
top_suppliers_sp = pivot_sp.sum(axis=1).sort_values(ascending=False).head(10).index
pivot_sp_top10 = pivot_sp.loc[top_suppliers_sp]

# 3. Heatmap ile çiz
plt.figure(figsize=(10, 6))
sns.heatmap(pivot_sp_top10, annot=True, fmt='d', cmap="BuGn")
plt.title("Tedarikçilere Göre Sahte Parça (Fiziksel) Kontrol Sonuçları (İlk 10)")
plt.xlabel("Kontrol Sonucu")
plt.ylabel("Tedarikçi")
plt.tight_layout()
plt.show()
```

```python
# En basit haliyle: Sahte Parça (Fiziksel) kontrol sonuçlarının toplam sayısını gösteren heatmap

# 1. Sahte Parça (Fiziksel) değerlerini say
simple_counts = df["Sahte Parça (Fiziksel)"].value_counts().to_frame()

# 2. Kolon adını anlamlı yap
simple_counts.columns = ["Kayıt Sayısı"]

# 3. Heatmap çiz
plt.figure(figsize=(4, 3))
sns.heatmap(simple_counts, annot=True, fmt='d', cmap="YlGnBu")
plt.title("Sahte Parça (Fiziksel) Kontrol Sonuçları")
plt.ylabel("Sonuç")
plt.xlabel("")
plt.tight_layout()
plt.show()
```

```python
# 1) Tarih sütununu gün seviyesine düşür
df["Kayıt Tarihi"] = pd.to_datetime(df["Kayıt Tarihi"], dayfirst=True, errors="coerce")
df["Gün"] = df["Kayıt Tarihi"].dt.date

# 2) Günlük uygun / uygun değil pivot-tablosu
gunluk_full = (df
               .groupby(["Gün", df["Denetim Sonucu"].str.upper()])
               .size()
               .unstack(fill_value=0)
               .rename(columns={"UYGUN": "Uygun", "UYGUN DEĞİL": "Uygun Değil"})
              )

# 3) Kapsamlı tarih aralığı oluştur → eksik günleri sıfırla
tam_aralik = pd.date_range(df["Gün"].min(), df["Gün"].max()).date
gunluk_full = (gunluk_full
               .reindex(tam_aralik, fill_value=0)
               .rename_axis("Tarih")
               .reset_index())

# 4) Text rapor bloğu
print("GÜNLÜK KAYIT DAĞILIMI (Tüm veri aralığı)")
print("-" * 50)
print(f"{'Tarih':<12} {'Uygun':>7} {'Uyg.Deg.':>10}")
print("-" * 50)
for _, row in gunluk_full.iterrows():
    print(f"{row['Tarih']:<12} {row['Uygun']:>7} {row['Uygun Değil']:>10}")
```

```python
# Tarih kolonunu önce datetime64 → sonra strftime('%Y-%m-%d') string
gunluk_full["Tarih"] = (
    pd.to_datetime(gunluk_full["Tarih"])    # emin ol datetime
      .dt.strftime("%Y-%m-%d")              # '2025-07-22' formatına çevir
)
```

```python
# ---- 2) Tedarikçi detay tablosu ----------------------------------------
tedarikci_stats = (
    hafta_df
    .assign(Sonuc=hafta_df["Denetim Sonucu"].str.upper())
    .pivot_table(index="Tedarikçi",
                 columns="Sonuc",
                 values="Kayıt Tarihi",
                 aggfunc="count",
                 fill_value=0)
    .rename(columns={"UYGUN": "Uygun", "UYGUN DEĞİL": "Uygun Değil"})
)
tedarikci_stats["Toplam Kayıt"] = tedarikci_stats.sum(axis=1)
tedarikci_stats = tedarikci_stats.sort_values("Toplam Kayıt", ascending=False)

# ---- 3) Malzeme detay tablosu -----------------------------------------
hafta_df["Parti_Toplam"] = hafta_df["Parti Adedi"].fillna(0) * hafta_df["Parti Miktarı"].fillna(0)

malzeme_stats = (
    hafta_df
    .assign(Sonuc=hafta_df["Denetim Sonucu"].str.upper())
    .pivot_table(index="Malzeme Tanımı",
                 columns="Sonuc",
                 values="Kayıt Tarihi",
                 aggfunc="count",
                 fill_value=0)
    .rename(columns={"UYGUN": "Uygun", "UYGUN DEĞİL": "Uygun Değil"})
)

# Parti toplamı ekle
malzeme_stats["Parti_Toplam"] = (
    hafta_df.groupby("Malzeme Tanımı")["Parti_Toplam"].sum()
)
malzeme_stats["Toplam Kayıt"] = malzeme_stats[["Uygun","Uygun Değil"]].sum(axis=1)
malzeme_stats = malzeme_stats.sort_values("Toplam Kayıt", ascending=False)

# ---- En çok / en az -----------------------------------------------------
top_tedarikci  = tedarikci_stats.head(1).index[0]
low_tedarikci  = tedarikci_stats.tail(1).index[0]
top_malzeme    = malzeme_stats.head(1).index[0]
low_malzeme    = malzeme_stats.tail(1).index[0]

print("En çok kayıt girilen tedarikçi:", top_tedarikci)
print("En az kayıt girilen tedarikçi :", low_tedarikci)
print("En çok kayıt girilen malzeme   :", top_malzeme)
print("En az kayıt girilen malzeme    :", low_malzeme)
```
